```python
import pandas as pd
```

```python
df=pd.read_csv("Resume.csv")
df.head(20)
```

```python
import re
def cleanResume(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText)  # remove URLs
    resumeText = re.sub('RT|cc', ' ', resumeText)  # remove RT and cc
    resumeText = re.sub('#\S+', '', resumeText)  # remove hashtags
    resumeText = re.sub('@\S+', '  ', resumeText)  # remove mentions
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resumeText)  # remove punctuations
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText) 
    resumeText = re.sub('\s+', ' ', resumeText)  # remove extra whitespace
    return resumeText
    
df['cleaned_resume'] = df.Resume_str.apply(lambda x: cleanResume(x))
df.head(20)
```

```python
l=['Resume_str','Resume_html']
df=df.drop(l,axis=1)
df.head()
```

```python
# Step 1: Data Preparation (Assuming 'cleaned_data' contains labeled resumes)
# ...

# Step 2: Feature Engineering (TF-IDF Vectorization)
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['cleaned_resume'])

# Step 3: Model Training (Assuming 'classifier' is the trained model)
from sklearn.svm import SVC  # Example classifier
classifier = SVC(kernel='linear')
X = tfidf_matrix  # Features
y = df['Category']  # Target variable
classifier.fit(X, y)

# Step 4: Prediction
job_description = "We need teachers "
job_desc_vectorized = vectorizer.transform([job_description])

# Step 5: Model Application (Predict Relevant Category)
predicted_category = classifier.predict(job_desc_vectorized)[0]
#print(predicted_category)

# Step 6: Retrieve Relevant Resumes
relevant_resumes = df[df['Category'] == predicted_category]
print(relevant_resumes)
```
